```python
import pandas as pd
# data = pd.read_csv('/home/nesma/Desktop/Results/CIFAR10/Keras/Run1/log.txt', delim_whitespace=True)
# data
count = len(open("/home/nesma/Desktop/Results/CIFAR10/Keras/Run1/log.txt").readlines(  ))

lines= list(range(3,count-5,2)) 
# [3, count-5]


i=0
linesList = []
linesList.append([5,0])
f=open('/home/nesma/Desktop/Results/CIFAR10/Keras/Run1/log.txt')
for line in f:
    if i in lines:
        tempList = line.split('-')[4:]
        tempList[0] = tempList[0].split(':')[1]
        tempList[1] = tempList[1].split(':')[1].replace('\n','')
        
        linesList.append(tempList)
    i+=1
    
newData = pd.DataFrame(linesList)
newData.columns = ['val_loss','val_acc']  

newData.dtypes
newData.val_acc = pd.to_numeric(newData.val_acc, errors='ignore')
newData.val_loss = pd.to_numeric(newData.val_loss, errors='ignore')
newData.val_loss = (newData.val_loss-newData.val_loss.min())/(newData.val_loss.max()-newData.val_loss.min())
newData.val_loss = newData.val_loss*10

```
